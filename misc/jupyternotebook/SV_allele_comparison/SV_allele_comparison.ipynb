```python
#!~/.conda/envs/lra/bin/python3
import os
import sys
import re
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from collections import defaultdict 
from matplotlib.collections import PathCollection
import numpy as np
import random

os.chdir(r"/project/mchaisso_100/cmb-16/jingwenr/trfCall/revision/sv-vntr/vamos/combineAll/individual/")
```

```python
compressions = ["original", "q-0.1", "q-0.2", "q-0.3"]
```

```python
data_dict= {"compression": [], "sv_allele": [], \
                          "sv_len_allele": [], \
                          "vamos_len_allele": [],\
                          "vamos_allele": []}
```

```python
for compression in compressions:
    file_name = f"{compression}/indel20/{compression}.all.bed"
    with open(file_name, 'r') as fin:
        lines = fin.readlines()
        for line in lines:
            line = line.rstrip('\n')
            fields = line.split('\t')
            vntr = fields[0]
            sv_allele = int(fields[1])
            sv_len_allele = int(fields[2])
            vamos_len_allele = int(fields[3])
            vamos_allele = int(fields[4])
            
            data_dict['sv_allele'].append(sv_allele)
            data_dict['sv_len_allele'].append(sv_len_allele)
            data_dict['vamos_len_allele'].append(vamos_len_allele)
            data_dict['vamos_allele'].append(vamos_allele)
            data_dict['compression'].append(compression)
```

```python
df = pd.DataFrame.from_dict(data_dict)
```
