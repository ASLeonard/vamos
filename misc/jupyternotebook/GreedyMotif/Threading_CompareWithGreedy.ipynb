```python
def readVCF(in_file, asm_dict, asm):
    print(f'start: {asm}')
    with open(in_file, 'r') as fin:
        lines = fin.readlines()
        for idx, line in enumerate(lines):
            if line.startswith('#'): continue
            fields = re.split('\t', line.strip('\n'))
            chrom, start, end = fields[0], fields[1], fields[2]
            motif_list = list(fields[3].split(','))
            infos = re.split(':', re.split(';', fields[4])[0])
            read_name = infos[0]
            anno_index = [int(i) for i in infos[3].split(',')]
            anno_str = ''.join([motif_list[i] for i in anno_index])
            read_seq = infos[4]
            d = lv.distance(read_seq, anno_str)
            asm_dict[asm] = {}
            asm_dict[asm][chrom + "_" + start + '-' + end] = {'motif': motif_list, 'annoIndex': anno_index, 'annoStr': anno_str, 'seq': read_seq, 'dist': d}
#             asm_dict[asm][chrom + "_" + start + '-' + end]['motif'] = motif_list
#             asm_dict[asm][chrom + "_" + start + '-' + end]['annoIndex'] = anno_index
#             asm_dict[asm][chrom + "_" + start + '-' + end]['annoStr'] = anno_str
#             asm_dict[asm][chrom + "_" + start + '-' + end]['seq'] = read_seq
#             asm_dict[asm][chrom + "_" + start + '-' + end]['dist'] = d
    print("finish: ", asm)
```
